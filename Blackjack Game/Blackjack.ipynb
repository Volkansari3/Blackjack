```python
from random import shuffle
```

```python
suits = ["Spades", "Diamonds" , "Clubs", "Hearts"]
```

```python
ranks = ["Two", "Three", "Four", "Five", "Six", "Seven", "Eight", "Nine", "Ten","Jack", "Queen", "King", "Ace"]
```

```python
values = {"Two": 2, "Three": 3, "Four": 4, "Five": 5, "Six": 6, "Seven": 7, "Eight": 8, "Nine": 9, "Ten": 10,
          "Jack": 10, "Queen": 10, "King": 10, "Ace": 11}
```

```python
class Card():

    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]

    def __str__(self):
        return f"{self.rank} of {self.suit}"
```

```python
class Deck():

    def __init__(self):
        self.all_cards = []

        for suit in suits:
            for rank in ranks:
                single_card = Card(suit,rank)
                self.all_cards.append(single_card)

    def shuffle(self):
        shuffle(self.all_cards)

    def draw_card(self): 
        return self.all_cards.pop()
   
```

```python
class Player():

    def __init__(self,name):
        self.name = name
        self.balance = 2000
        self.bet = 0
        self.total = 0
        self.player_hand = []
        self.new_hand = []

    def place_bet(self):
        temp = True

        while temp:  
            try:
                new_bet = int(input("Please enter a bet: "))
            
                if new_bet > self.balance:
                    print("Insuficient balance please try again!")
                    
                elif new_bet <= 0:
                    print("Bet amount must be greater than 0!")

                else:
                    self.bet = new_bet
                    self.balance -= self.bet
                    print(f"The bet is: {self.bet}")
                    print(f"Remaining balance is: {self.balance}\n")
                    temp = False

            except:
                print("Invalid input please try again!")
            
    def Hit(self,new_card):
        self.player_hand.append(new_card) 
        print(f"{self.name} drew a {new_card}")

    def Stand(self):
        pass

    def Double_Down(self,new_card):
        self.bet = self.bet * 2
        print(f"Your new bet is: {self.bet}")   
        print(f"{self.name} draws a card")
        self.player_hand.append(new_card) 

    def Split(self,card1,card2):  
        print("Your hand is splitting...")  
        self.new_hand = [self.player_hand.pop()]
        print(f"First hand: {self.new_hand[0]}")
        print(f"Second hand: {self.player_hand[0]}\n")

        print("One card is drawn for the first hand and one for the second hand.")
        self.player_hand.append(card1)
        self.new_hand.append(card2)

        print(f"First hand: {[str(x) for x in self.player_hand]}")
        print(f"Second hand: {[str(y) for y in self.new_hand]}")

    def Surrender(self):
        self.bet = self.bet / 2
        self.balance = self.balance + self.bet
        print(f"Player {self.name} is surrendering")
        
    def Player_Point(self):   
        if len(self.player_hand) < 2:
            print("Error: Player does not have enough cards!")

        else:
            self.total = 0
            ace_counter = 0
            
            for cards in self.player_hand:        
                if cards.rank == "Ace":   
                    self.total += 11
                    ace_counter += 1
                else:
                    self.total += cards.value

            while self.total > 21 and ace_counter > 0:
                self.total -= 10
                ace_counter -= 1
                
        return self.total
```

```python
class Dealer():

    def __init__(self):
        self.dealer_hand = []
        self.total = 0

    def Deal_card(self,new_cards):
        self.dealer_hand.append(new_cards) 

    def Show_Cards(self,reveal):
        if reveal == True:
            print(f"Dealer hand is: {[str(list2) for list2 in self.dealer_hand]}\n")  

        else:
            print(f"Dealer hand is: {self.dealer_hand[1]} and the second card is hidden")

    def Dealer_Point(self):
        if len(self.dealer_hand) < 2:
            print("Error: Not enough cards for the dealer!")

        else:
            self.total = 0
            ace_counter = 0
            
            for cards in self.dealer_hand:
                if cards.rank == "Ace":
                    self.total += 11
                    ace_counter += 1
                else:
                    self.total += cards.value

            while self.total > 21 and ace_counter > 0:
                self.total -= 10
                ace_counter -= 1
                
        return self.total       
```
